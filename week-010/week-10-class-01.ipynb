# DATABASES

#SQL

RDBMS Relational database management systems 


*   1970's 
*  RDBMS dominant method for storing **structured** data in web and business applications





ACID


* *Atomic* 
    * unit of work which must be completed with *all* of its data modifications
* *Consistent* 
    * at end of the transaction -- all data must be left in a consistent state
* *Isolated* 
    * modifications of data must be independent of another transaction
* *Durable* 
    * effects of the modifications performed by the transaction must be permanent

Distributed Systems

* Network of mainframes, workstations, personal 
computers, printers wherein the computers interact with each other and share the resources of the system to achieve a common goal

ADVANTAGES
* Reliability (fault tolerance) 
    * if some of the machines within the system crash/go offline --  the rest of the computers remain unaffected 
* Scalability 
    * expand by adding more machines
* Speed/Performance 
    * collection of processors = more computing power and speed 

DISADVANTAGES
* Troubleshooting 
    * diagnosing problems on network is tricky
* Network 
    * typical infrastructure comes with problems such -- transmission problems, overloading etc
* Security 
    * sharing of data on network makes it easier to access but, increases security risk 


SCALE 
* Ability of system to expand

Vertical scaling
* *scale up* - add resources within a system to increase capacity
    * add CPUs to an existing server
    * increase memory 
    * expand storage by adding hard drive
    * **NOTE**: adding hardware is expensive

Horizontal scaling
* *scale* out - add more nodes to a system
    * add new computer to distributed software appliction
    * NoSQL system takes advantage of “scaling out” add more nodes to a system and distribute the load over those nodes
    * **NOTE** this is more cost effect when dealing with HUGE volumes of data


#NoSQL

NoSQL
* non-relational database management system 
* Emerged late 2000s


Why NoSQL
* exponential increase in data capture:
    * personal user information, social graphs (FB), geo location data(maps), user-generated content (yelp) and machine logging data (they are watching you;) 
* along with increase comes variety: 
    * structured, semi-structured, and polymorphic —
defining the schema (like SQL) in advance became nearly impossible
    * NoSQL databases allow developers to store huge amounts of unstructured data -- giving them a lot of flexibility
* cloud computing
    * developers using public clouds to host applications and 
data 
    * ability to distribute data across multiple servers and regions -- makes data resilient - to scale out instead of scale up



RDBMS vs NoSQL
* RDBMS
    * Structured query language (SQL)
    * Data and its relationships are stored in separate tables/entities
    * Each table has a predefined Schema - column with be int/char / length etc
    * Language: Select * FROM Name ORDER BY 

* NoSQL
    * Not Only SQL / NoSQL
    * No declarative query language
    * No predefined schema


**CAP (Brewer’s Theorem)**
* 3 basic requirements in distributed architecture 

**Consistency**
* data in the database remains consistent after the execution of an operation
    * exp after an update operation all clients see the same data
    
**Availability** 
* system is always on -- no downtime

**Partition Tolerance** 
* system continues to function even when part of it goes down
Exp set up groups

basically impossible to fill all 3 CAP requirements in on system
* most NoSQL database follow t2/3  

* CA 
    * Single site cluster, all nodes are always in contact When a partition occurs, the system blocks
* CP 
 * Some data may not be accessible, but the rest is still consistent/accurate
* AP 
 * System is still available under partitioning, but some of the data returned may be inaccurate


#TYPES OF NoSQL DATABES
* Key 
* Column
* Document
* Graph


Key-Value pair storage 
* basic form of NoSQL databases
* designed to handle huge amounts of data
* key value stores schema-less data 
* database stores data as hash table 
    * each key is unique
    * value can be string, JSON, BLOB (Binary Large OBjec) etc.
    * key may be strings, hashes, lists, sets, sorted sets 
    * values are stored against these keys
        * example  
            * key: "Name" 
            * associated value: "Keino"
            * Key-Value stores can be used as collections, dictionaries, associative arrays etc
* Key-Value stores follow the *'Availability'* and *'Partition'* aspects of CAP theorem
* key value stores are great for: 
    * *shopping cart contents* 
    * individual values like color schemes
    * default account number
*examples 
    * Redis, Dynamo, Riak.





Column-oriented databases
* every column is treated individually.
* values of a single column are stored contiguously
* column stores data in column specific files
* query processors work on columns 
* all data within each column datafile have the same type which makes it ideal for compression
    * format improves performance of queries - by accessing specific column data
    * high performance on aggregation queries (e.g. COUNT, SUM, AVG, MIN, MAX).
* good for: 
    * data warehouses and business intelligence, customer Relationship management (CRM), Library card catalogs etc.
* Example 
 * BigTable, Cassandra, SimpleDB etc.


Graph databases
* collection of nodes and edges
 * each node -- represents an entity 
 * each edge -- represents relationship between two nodes
    * nodes and edges are defined by a unique identifier
* as the number of nodes increases, the cost of a local step (or hop) remains the same


Document Oriented databases
* data in this model is stored inside documents
    * database is collection of documents
* document is a key value collection where the key allows access to its value
* documents are not typically forced to have a schema and therefore are flexible and easy to change
* documents are stored into collections in order to group different kinds of data
* documents can contain many different key-value pairs, or key-array pairs, or even nested documents.
