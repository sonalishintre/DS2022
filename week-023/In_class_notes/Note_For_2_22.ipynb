# Review of Pandas


    There are two default Data Structures in Pandas
    
    
    Series: Homogenous and Indexed 
    Dataframe: Created from multiple Series
    
    
    df.head() -- return top N rows from a Dataframe
    
    df.tail() -- return the bottom N rows from a Dataframe
    
    df[col_name] -- return the specified Series
    
    df.shape -- returns some data about our dataframe, gives n dimesions
   
    .isin -- filter using a range or iterable || using or
    
    df[column_name] condition -- returns all the rows where the condition is met
    
    df.notna() -- returns the values that aren't Null 
    
    df.loc[what_rows,what_columns] -- create a subsection dataframe
    
    df.columns -- returns all the column names
    
    df.iloc[what_rows, what_columns]

# How can we quickly create derived columns?

```python
air_quality["london_mg_per_cubic"] = air_quality["station_london"] * 1.882
air_quality
```

```python
air_quality["ratio_paris_antwerp"] = air_quality["station_paris"] / air_quality["station_antwerp"]
air_quality
```

```python
# Renaming columns

air_quality.rename(columns={"station_antwerp":"BT3003"})
```

```python
#Some Useful things to know
# Replace in the column name


air_quality.columns = air_quality.columns.str.replace("_","!")
air_quality.columns = air_quality.columns.str.upper()
```

```python
titanic
```

# Grouping in Pandas

    Pandas uses the Split-Apply-Combine Design pattern

```python
titanic[["Sex","Age"]].groupby("Sex")
```

```python
titanic[["Sex","Age"]].groupby("Sex").mean()
```

```python
titanic.groupby("Sex").mean()
```

```python
titanic.groupby("Sex")["Age"].mean()
```

```python
titanic.groupby(["Sex","Pclass"])["Fare"].mean()
```

```python
#Return things as a percentage of the whole
titanic["Pclass"].value_counts() / titanic["Pclass"].value_counts().sum() *100
```

```python
titanic.groupby(["Pclass","Sex"])["Pclass"].count()
```

```python
air_quality = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/air_quality_long.csv", index_col="date.utc", parse_dates=True)
```

```python
no2 = air_quality[air_quality["parameter"] == "no2"]
no2
```

```python
no2.pivot(columns="location",values="value").plot()
```

```python
air_quality.pivot_table(
    values="value",
    index="location",
    columns="parameter",
    aggfunc="mean",
    
)
```

```python
air_quality.pivot_table(
    values="value",
    index="location",
    columns="parameter",
    aggfunc="median",
    margins=True
)
```
