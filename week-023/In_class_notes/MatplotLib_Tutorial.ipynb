# MatPlotLib
    
    We have had a ton of practice now in Pandas, in shaping our Data, in
    creating new rows, and new columns, and new tables, and all of this is 
    useful. In a vacuum it might seem like busy work, but one of the end 
    results of all these transformations and all of these shapings is 
    visualization. This is a way to display our data in a more effective
    way than just simple numbers. It's not that numbers are hard to understand
    but without seeing the "bigger" picture being told by our data, it can
    be hard to determine the next steps.
    
    Pandas does not actually handle visualization. Pandas is once again
    only for the DataFrame and Series type logic. We use Matplotlib as our
    visualization library. Matplotlib has a ton of utlity, and a ton of power
    so it's important that we go over the library so that we can better use
    it.
    

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html#matplotlib.pyplot.subplots

```python
fig,ax = plt.subplots() # Create a figure containing a single Axes

ax.plot([1,2,3,4],[1,2,3,4]) # We are passing a list of X values, and a Series
                            # of Y values.

```

```python
fig = plt.figure()  # an empty figure with no Axes
fig, ax = plt.subplots()  # a figure with a single Axes
fig, axs = plt.subplots(2, 1)  # a figure with a 2x2 grid of Axes
```

https://matplotlib.org/stable/api/figure_api.html#matplotlib.figure.Figure

# The Figure Object

    The figure object can be seen as our outermost wrapper. It contains all of the "higher" order objects, such as the
    Axes themselves, Legends, Title and so on. It is a useful wrapper that allows us to seperate logic within the
    visualizations.

```python
fig = plt.figure()
```

```python
fig.suptitle("Evil",fontsize=16)
```
