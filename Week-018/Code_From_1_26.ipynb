```python
import random
import pandas as pd
from matplotlib import pyplot as plt
```

```python
rand_numbers = [random.randrange(1,100) for i in range(100)]
```

```python
plt.hist(rand_numbers,bins=100)
plt.show()
```

```python
df = pd.DataFrame(rand_numbers)
df
```

```python
population_variance = int(df.var(ddof=0))
population_variance
```

```python
df_sample_size_10 = df.sample(50).reset_index(drop=True)
df_sample_size_10
df_sample_size_10.var(ddof=0)
```

```python
variance_of_samples=[]
mean_rolling_variance = []

for i in range(1,101):
    
    sample_of_10 = df.sample(50).reset_index(drop = True)
    sample_of_10_variance = int(sample_of_10.var(ddof=0))
    
    variance_of_samples.append(sample_of_10_variance)
    
    if(i == 1):
        mean_rolling_variance.append(sample_of_10_variance)
    else:
        mean_rolling_variance.append(sum(variance_of_samples)/i)
```

```python
df_100_samples = pd.DataFrame(list(zip(variance_of_samples,mean_rolling_variance)),
                              columns=["Sample Variance","Rolling Variance"])

```

```python
fig,ax = plt.subplots()
df_100_samples.reset_index().plot.scatter(x="index",y="Rolling Variance",ax=ax)
ax.plot([0,100],[population_variance,population_variance],linestyle="--",color="red")
```

```python
fig1,ax1 = plt.subplots()
df_100_samples.reset_index().plot.scatter(x="index",y="Sample Variance",ax=ax1)
ax1.plot([0,100],[population_variance,population_variance],linestyle="--",color="red")
```

```python
fig3,ax3 = plt.subplots()
df_100_samples.drop(["Rolling Variance"],axis=1).plot.kde(ax=ax3)
ax3.plot([population_variance,population_variance],[0,.004],linestyle="--",color="red")
```
