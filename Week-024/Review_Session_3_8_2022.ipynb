```python
# Selection Sort

def selection_sort(arr):
    n = 0
    while len(arr[n:]) > 1:
        i=0+n
        j=1+n
        flips = 0
        while j <= len(arr)-1:
                if arr[i] > arr[j]:
                    arr[i],arr[j] = arr[j],arr[i]
                    flips+=1
        j+=1
        if flips==0:
            return arr
    n+=1
    return arr
    
                
```
