```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_validate
```

```python
df = pd.read_csv('kc_house_data.csv')
```

```python
df
```

# Let's do a simple holdout

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop(columns='price'),df['price'],test_size =0.3)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
to_train_x = x_train[['sqft_living','floors']]
```

```python
to_train_x
```

```python
regressor = LinearRegression()
scores = cross_val_score(regressor,to_train_x,y_train,cv=5)
print(scores)
print(scores.mean())
```

```python
regressor = LinearRegression()
scores = cross_val_score(regressor,to_train_x,y_train,cv=5,scoring='neg_mean_squared_error')
print(scores)
print(-(scores.mean()))
```

```python
regressor = LinearRegression()
scores = cross_val_score(regressor,to_train_x,y_train,cv=ShuffleSplit(n_splits=5,test_size=0.3),scoring='neg_mean_squared_error')
print(scores)
print(-(scores.mean()))
```

```python
regressor = LinearRegression()
scores = cross_validate(regressor,to_train_x,y_train,cv=ShuffleSplit(n_splits=5,test_size=0.3),scoring=('neg_mean_squared_error','r2'))
scores['test_neg_mean_squared_error']
scores['test_r2']
```

# How can we use this to be exhaustive?

```python
test_with = ['sqft_living','floors']
```

```python
for i in range(1,len(test_with)+1):
    to_train_x = x_train[test_with[0:i]]
    regressor = LinearRegression()
    scores = cross_validate(regressor,to_train_x,y_train,cv=ShuffleSplit(n_splits=5,test_size=0.3),scoring=('neg_mean_squared_error','r2'))
    print(f'For the features {to_train_x.columns}')
    print(f'The MSE scores were {-scores["test_neg_mean_squared_error"]}')
    print(f'The r2 scores were {scores["test_r2"]}')
    print(f'The mean MSE scores was {-scores["test_neg_mean_squared_error"].mean()}')
    print(f'The mean r2 score was {scores["test_r2"].mean()}')
    print('\n')
    
    
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
test_with = [LinearRegression(),SGDRegressor()]
```

```python
for i in range(0,len(test_with)):
    to_train_x = x_train[['sqft_living','floors']]
    regressor = test_with[i]
    scores = cross_validate(regressor,to_train_x,y_train,cv=ShuffleSplit(n_splits=5,test_size=0.3),scoring=('neg_mean_squared_error','r2'))
    print(f'For the Regressor {regressor}')
    print(f'The MSE scores were {-scores["test_neg_mean_squared_error"]}')
    print(f'The r2 scores were {scores["test_r2"]}')
    print(f'The mean MSE scores was {-scores["test_neg_mean_squared_error"].mean()}')
    print(f'The mean r2 score was {scores["test_r2"].mean()}')
    print('\n')
    
```
